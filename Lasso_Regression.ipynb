```python
# üì¶ Import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import Lasso

# üß™ Create dummy data
X = np.array([[1], [2], [3], [4], [5], [6],[5], [6]])
Y = np.array([[2], [4], [5], [4], [5], [7],[5], [6]])

# ü§ñ Create and train Lasso Regression model
model = Lasso(alpha=0.1)  # You can tune alpha to control regularization
model.fit(X, Y)

# üìà Predict values
Y_pred = model.predict(X)

# üñºÔ∏è Plot actual vs predicted
plt.scatter(X, Y, color='red', label='Actual Data')
plt.plot(X, Y_pred, color='blue', linewidth=2, label='Lasso Prediction')
plt.title('Lasso Regression on Dummy Data')
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.legend()
plt.grid(True)
plt.show()
```
